```python
#to open file for read operation
filev = open("firstfile.txt", "r")
# to check file is open or not
if(filev):
    #to read dat from file
    data = filev.read()
    print(data)
    print("------------------------")
    print("no of characters :",len(data))
    #closing file
    filev.close()
else:
    print("unable to open the file")
```
