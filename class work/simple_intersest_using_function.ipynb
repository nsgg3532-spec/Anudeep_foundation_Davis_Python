```python
def calculatesi(p,r,t):
  return ((p*r*t)/100)
principal = float(input("enter principal (in Rs):"))
rate = float(input("enter principal (in %):"))
time = float(input("enter principal (in year):"))
print("simple interest :rs",calculatesi(principal,rate,time))
```
