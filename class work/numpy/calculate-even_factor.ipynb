```
import numpy as np

n = float(input("enter even factor"))
if(n<0):
  print("negative value is not allow")
else:
 arr = np.arange(1, n + 1)
 even_factors = arr[(n % arr == 0) & (arr % 2 == 0)]

 print(even_factors)
```
