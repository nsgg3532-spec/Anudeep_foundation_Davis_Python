```python
# to open file for read operation
filev = open("firstfile.txt", "r")

# to check file is open or not
if(filev):
    # to read data from file
    data = filev.read()
    print(data)
    print("------------------------")
    # vowels list (both uppercase and lowercase)
    vowels = "aeiouAEIOU"
    count = 0
    # loop to check each character
    for ch in data:
        if ch in vowels:
            count += 1
    # to print number of vowels
    print("number of vowels :", count)
    # closing file
    filev.close()

else:
    print("unable to open the file")
```
