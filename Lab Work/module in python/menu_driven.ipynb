```python
import threedfigures

print("1. Cube")
print("2. Cuboid")
print("3. Cylinder")
print("4. Cone")
print("5. Sphere")

choice = int(input("Enter your choice: "))

print("1. Curved Surface Area")
print("2. Total Surface Area")
print("3. Volume")

op = int(input("Enter operation: "))

if choice == 1:
    a = float(input("Enter side: "))
    if op == 1:
        print(threedfigures.cube_csa(a))
    elif op == 2:
        print(threedfigures.cube_tsa(a))
    else:
        print(threedfigures.cube_volume(a))

elif choice == 2:
    l = float(input("Length: "))
    b = float(input("Breadth: "))
    h = float(input("Height: "))
    if op == 1:
        print(threedfigures.cuboid_csa(l,b,h))
    elif op == 2:
        print(threedfigures.cuboid_tsa(l,b,h))
    else:
        print(threedfigures.cuboid_volume(l,b,h))

elif choice == 3:
    r = float(input("Radius: "))
    h = float(input("Height: "))
    if op == 1:
        print(threedfigures.cylinder_csa(r,h))
    elif op == 2:
        print(threedfigures.cylinder_tsa(r,h))
    else:
        print(threedfigures.cylinder_volume(r,h))

elif choice == 4:
    r = float(input("Radius: "))
    l = float(input("Slant height: "))
    h = float(input("Height: "))
    if op == 1:
        print(threedfigures.cone_csa(r,l))
    elif op == 2:
        print(threedfigures.cone_tsa(r,l))
    else:
        print(threedfigures.cone_volume(r,h))

elif choice == 5:
    r = float(input("Radius: "))
    if op == 1 or op == 2:
        print(threedfigures.sphere_tsa(r))
    else:
        print(threedfigures.sphere_volume(r))
```
